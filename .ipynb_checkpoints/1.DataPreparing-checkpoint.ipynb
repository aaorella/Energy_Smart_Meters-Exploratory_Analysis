# Corrigiendo mediciones malas
en ocaciones hay mmediciones que dicen por ejemplo 18 pero el verdadero valor es 180

```python
for_update_df = ((df_join[["voltajeA", "voltajeB", "voltajeC", "angvoltajeB", "angvoltajeC"]]).applymap(lambda x: x if (x>=10) else x*10))
```

```python
df_join.update(for_update_df)
```

## Eliminado mediciones erroneas de filas de 0

## Obteniendo el consumo por medicion
se usar .diff

```python
df_join["kw_consumido"] = df_join.groupby("Medidor")["totalkwh"].transform(lambda x: x.diff())
```

```python
df_join.fillna(0, inplace =True)
df_join
```

tras revisar las mediciones de estos medidores nos pudimos dar cuenta que el contador de Totalkwh se regresa a cero cuando alcanza el valor de 999999

```python
def correct(serie):
    serie = serie.reset_index()
    #serie.columns = ["1","kw_consumido"]
    print(serie.columns)
    ids = (serie.index[serie["kw_consumido"] <0])
    if (len(ids) >0):
        #print(serie.columns)
        #print(ids)
        #print(len(ids))
        for i in ids:
            #print(i)
            print(serie.iloc[i])
            print(serie.iloc[i -1])
            serie.iloc[i,26] = serie.iloc[i,18] + 1000000 - serie.iloc[i -1, 18]
    return serie.drop(columns=['index'])
```

```python
df_join_grouped = df_join.groupby("Medidor").apply(correct)
```

```python
df_limpia = df_join_grouped.drop(columns=["Medidor"]).reset_index()
```

```python
df_limpia = df_limpia.drop(columns=["level_1"])
```
