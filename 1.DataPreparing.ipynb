## Adding kwh_consumido_hora

```python
def addKwh(data):
    datetime = pd.to_datetime(data["fecha"])
    datetime_diff = datetime.diff().dt.seconds / 3600
    data["kwh_hora"] = data["kwh_consumido"] / datetime_diff
    return data
```

## Getting the larger windows

```python
#df_f11.sort_values(by = "fecha", ascending = True, inplace = True)
```

```python
df_f11["fecha"] = pd.to_datetime(df_f11["fecha"])
```

```python
df_f11["dayMonth"] = df_f11["fecha"].dt.week
```

```python
list_datetime = df_f11["fecha"].dt.date.value_counts().index.sort_values().to_list()
list_datetime[:10]
```

```python
tail = list_datetime[-1]
head = list_datetime[0]
```

```python
dayUniques = (tail - head ).days
print("There is a total of {} days in our dataset".format(dayUniques))
```

```python
def countDaysMeters(data, dayUniques, head, tail):
    count = 0
    good_meters = []
    for name, group in data.groupby("medidor"):
        group.set_index("fecha", inplace = True)
        group = group[head:tail]
        group = group.resample("24H").mean()
        if(group["dayMonth"].value_counts().sum() >= dayUniques):
            if(not group.isnull().any().any()): #any value is different than nan
                good_meters.append(name)
                count +=1
    return count, good_meters
```

```python
def getMeters_maxDays(df, list_datetime, min_days = 0, direction = "left"):
    directions = ["left", "right", "both"]
    if(not(direction in directions)):
        return [], []
    
    #defining static edge
    if(direction == "left"):
        tail = list_datetime[-1]
    elif(direction == "right"):
        head = list_datetime[0]
    
    max_possible_offeset = len(list_datetime)
    if(direction == "both"):
        max_possible_offeset = len(list_datetime) // 2
        
    max_meters = []
    count_meters_max = 0
    head_and_tail = []
    for offset in  range(max_possible_offeset):
        #defining dynamic edges
        if(direction == "left" or direction == "both"):
            head = list_datetime[offset]
        if(direction == "right" or direction == "both"):
            tail = list_datetime[offset*(-1) - 1]
        
        num_days = (tail - head ).days
        
        if (num_days >= min_days):
            count_meters, meters = countDaysMeters(df_f11, num_days, head, tail)
            print("From {}  to {}  with {} days, we have {} meters".format(head, tail, num_days, count_meters))
            if(count_meters_max < count_meters):
                count_meters_max = count_meters
                max_meters = meters
                head_and_tail = [head, tail]
    return max_meters, head_and_tail
        
    
```

```python
meters1, dates1 = getMeters_maxDays(df_f11, list_datetime, min_days= 60, direction="left")
```

```python
meters2, dates2 = getMeters_maxDays(df_f11, list_datetime, min_days= 60, direction="right")
```

```python
meters3, dates3 = getMeters_maxDays(df_f11, list_datetime, min_days= 60, direction="both")
```

**note** we want a min two months, so min_days = 60

From the above functions we decided to select the windows from {2019-04-04} to {2019-04-04} with {112} meters
